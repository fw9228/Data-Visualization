# Scatter Plot

```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing built-in Data from Seaborn library

tips_data = sns.load_dataset('tips')

tips_data.head()
```

***Replot and Catplot***

Replot will be used for numerical data plotting and catplot will be used for categorical data plotting

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips_data)
```

***Hue, Style, and Size***

Categorical variables are good for this hue, style, and size


```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips_data, hue = 'smoker')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips_data, hue = 'smoker', style = "time")
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips_data, hue = 'smoker', style = 'time', size = 'size')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips_data, style = 'size')
```

```python
tips_data['size'].value_counts()
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips_data, style = 'smoker')
```

```python
tips_data['smoker'].value_counts()
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips_data, size = 'size', sizes = (15, 200))
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips_data, size = 'time', sizes = (15, 200))
```

```python
tips_data['time'].value_counts()
```

# Line Plot

```python
from seaborn import relplot

from numpy.random import randn
```

```python
#Creating a panda dataframe to use for the visualization

random_num = pd.DataFrame(dict(time= np.arange(600), value = randn(600).cumsum()))

random_num
```

```python
relplot(x = 'time', y = 'value', kind = 'line', data = random_num)
```

```python
#Importing built-in Data from Seaborn library

fmri_data = sns.load_dataset('fmri')
fmri_data.head()
```

```python
fig, ax = plt.subplots()
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri_data)
```

```python
relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri_data, label = 'Signal', height = 4, aspect = 2)
plt.legend(loc= 'right')
```

```python
#Without the confidence interval (ci)
relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri_data, ci = False, label = 'Signal', height = 4, aspect = 2)
plt.legend(loc= 'right')
```

```python
fmri_data[fmri_data['timepoint']==5]
```

```python
fmri_data[fmri_data['timepoint']==5].mean()
```

```python
fmri_data[fmri_data['timepoint']==5].min()
```

```python
#Without estimator
relplot(x = 'timepoint', y = 'signal', kind = 'line', estimator = None, data = fmri_data, label = 'Signal', height = 4, aspect = 2)
plt.legend(loc= 'right')
```

```python
#Adding a categorical variable (event) under hue
relplot(x = 'timepoint', y = 'signal', kind = 'line', hue = 'event', data = fmri_data, label = 'Signal', height = 4, aspect = 2)
plt.legend(loc= 'right')
```

```python
#Adding a categorical variable (event) under hue with marker being True
relplot(x = 'timepoint', y = 'signal', kind = 'line', hue = 'event', marker = True, data = fmri_data, label = 'Signal', height = 4, aspect = 2)
plt.legend(loc= 'right')
```

```python
#Adding two categorical variable (event and region) under hue and style
relplot(x = 'timepoint', y = 'signal', kind = 'line', hue = 'event', style = 'region', data = fmri_data, label = 'Signal', height = 4, aspect = 2)
plt.legend(loc= 'right')
```

```python
#Importing built-in Data from Seaborn library

sns.load_dataset('dots')
```

```python
#Interested in only dots data

dots_data = sns.load_dataset('dots').query("align == 'dots'")

dots_data.head()
```

```python
relplot(x = 'time', y = 'firing_rate', data = dots_data, kind = 'line')
```

```python
relplot(x = 'time', y = 'firing_rate', hue = 'coherence', data = dots_data, kind = 'line')
```

```python
relplot(x = 'time', y = 'firing_rate', hue = 'coherence', style = 'choice', data = dots_data, kind = 'line')
```

```python
#Change the line size
relplot(x = 'time', y = 'firing_rate', hue = 'coherence', style = 'choice', size = 'choice', data = dots_data, kind = 'line')
```

```python
#creating colors
adding_colors = sns.cubehelix_palette(light = 0.8, n_colors = 6)
relplot(x = 'time', y = 'firing_rate', hue = 'coherence', style = 'choice', data = dots_data, kind = 'line', palette = adding_colors)
```
